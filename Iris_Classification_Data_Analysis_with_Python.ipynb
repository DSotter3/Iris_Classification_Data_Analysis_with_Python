### 8. Split the data into training and testing data.  
   - Prepare your X and y, using appropriate variable names:  
- X: Drop the Species column  .
- y: Specify the Species colum  n   .
- Use sklearn train_test_split to split the d.  ta.

```python
X_train,X_test,y_train,y_test = train_test_split(iris.drop('Species',axis = 1), iris['Species'],test_size = 0.3)
```

### 9. Create the model and fit it to the training data.  
- Create an sklearn KNeighborsClassifier model, with k=1.  
- Use the fit method to fit it to the training data.  

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
```

### 10.Predict values based on testing data.  
◦ Use the predict method to predict values with the x testing data and store them 
in a variable。


```python
pred = knn.predict(X_test)
pred
```

### 11.Print out the classification report for the y test data and the predictions.

```python
print(classification_report(y_test,pred))
```

### 12. Use the knn.score method to print out a simplified score for the model.

```python
print('The accuracy is:',knn.score(X_test,y_test))
```

### 13. Repeat steps 9 to 12 using k=3, k=5, k=10, k=30, k=50.

#### When k=3:

```python
knn_3 = KNeighborsClassifier(n_neighbors =3)
knn_3.fit(X_train, y_train)
pred_3 = knn_3.predict(X_test)
print(pred_3)
print(classification_report(y_test,pred_3))
print('The accuracy is:',knn_3.score(X_test,y_test))
```

#### When k = 5:

```python
knn_5 = KNeighborsClassifier(n_neighbors=5)
knn_5.fit(X_train, y_train)
pred_5 = knn_5.predict(X_test)
print(pred_5)
print(classification_report(y_test,pred_5))
print('The accuracy is:',knn_5.score(X_test,y_test))
```

#### When k = 10:

```python
knn_10 = KNeighborsClassifier(n_neighbors=10)
knn_10.fit(X_train, y_train)
pred_10 = knn_10.predict(X_test)
print(pred_10)
print(classification_report(y_test,pred_10))
print('The accuracy is:',knn_10.score(X_test,y_test))
```

#### When k = 30:

```python
knn_30 = KNeighborsClassifier(n_neighbors=30)
knn_30.fit(X_train, y_train)
pred_30 = knn_30.predict(X_test)
print(pred_30)
print(classification_report(y_test,pred_30))
print('The accuracy is:',knn_30.score(X_test,y_test))
```

#### When k= 50:

```python
knn_50 = KNeighborsClassifier(n_neighbors=50)
knn_50.fit(X_train, y_train)
pred_50 = knn_50.predict(X_test)
print(pred_50)
print(classification_report(y_test,pred_50))
print('The accuracy is:',knn_50.score(X_test,y_test))
```
